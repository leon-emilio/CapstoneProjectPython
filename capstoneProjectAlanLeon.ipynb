```python
def loadSource(path):
    files = list()
    for file in os.listdir(path):

        parts = file.split(".")			#splits file name into array using the "." as a marker
        if len(parts) >= 2:				#if there are no dots, then the file doesn't even have extension
            tmp = ".".join(parts[:-1])	#everything except last element is sent to a temporary variable
            val = parts[-1] 			##Last element of the parts array is saved in val
            if val.lower() == "txt":	#.lower method to avoid using OR. Checks if the last element in the file name is "txt"
                files.append(file)		
    return files
```

## Regex Pattern to ignore punctuation marks

```python

markPattern = r'[^\w\s]'
```

```python
Times = []
starting = False
```

```python
for i in loadSource(r'C:\Users\tomat\OneDrive\Documentos\AmbientesCetys\Capstone'):
    with open('C:\\Users\\tomat\\OneDrive\\Documentos\\AmbientesCetys\\Capstone\\' +i, 'r', encoding='utf-8') as file:
        Times.append(i)
        start_time = time.time()
        for line in file:
            if '*** START OF THIS PROJECT GUTENBERG' in line:
                starting = True
                continue  # Skip the start line itself

            # Stop reading when the end line is found
            if '*** END OF THIS PROJECT GUTENBERG' in line:
                break  # Stop reading completely
                
            if starting:
            
                lineData = line.strip()
                lineData = lineData.lower()
                for char in lineData:
                    if re.match(markPattern, char) or char == '_':
                        continue
                    
                    try:
                        replace = replacements[char]
                        charCount[replace] += 1
                        
                    except KeyError:
                        charCount[char] += 1
        end_time = time.time() 
        time_taken = end_time - start_time
        Times.append(time_taken)
        starting = False
```

```python
top = sorted(charCount.items(), key=lambda item: item[1], reverse=True)


for key, value in top:
    print(f"{key} -> {value}")
```

```python
for i in range(0,len(Times) // 2):
    print(Times[i * 2] + " took " + f"{Times[i*2 + 1]} seconds" )
```
